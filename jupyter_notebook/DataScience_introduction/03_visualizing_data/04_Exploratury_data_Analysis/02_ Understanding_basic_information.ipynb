147개의 항목을 조사하얐고 997명이 참여하였다(147 columns,997 rows)

columns의 소개  
   - 0~18 : 음악 취향  
   - 19~30 : 영화 취향  
   - 31~62 : 취미/관심사  
   - 63~72 : 공포증  
   - 73~75 : 건강 습관  
   - 76~132 : 성격, 인생관 등  
   - 133~139 : 소비 습관  
   - 140~146 : 기본 정보  

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/young_survey.csv')
df.head()
```

기본정보를 파악해보자 140 이후를 다 가져오자

```python
basic_info=df.iloc[:,140:]
basic_info
```

```python
basic_info.describe()
```

참여자의 나이는 15세부터 30살이 참여했고 20대 초반이 주로 많다. 나머지 문자관련해서는 없다

문자 관련은 직접한번 보도록 하자

```python
basic_info['Gender'].value_counts()
```

여자가 조금더 많이 참여했다

```python
basic_info['Handedness'].value_counts()
```

오른손잡이가 훨 많다

```python
basic_info['Education'].value_counts()
```

최종학력은 고졸이 가장 많다.

그래프를 이용해서도 한번 알아보자

```python
sns.violinplot(data=basic_info, y='Age')
```

주로 10대 후반과 20대 초반의 참여자가 제일 많다는것을 알수 잇다

성별에따른 나이분포는 어케 될까?

```python
sns.violinplot(data=basic_info, y='Age',x='Gender')
```

오른손 잡이 왼손잡이에 따른 연령분포도 봐보도록 하자

```python
sns.violinplot(data=basic_info, y='Age',x='Gender',hue='Handedness')
```

키랑 몸무게에 대해서도 한번 보자

```python
sns.jointplot(data=basic_info, x='Height',y='Weight')
```
