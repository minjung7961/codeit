```python
import pandas as pd
```

```python
df = pd.read_csv('data/body.csv', index_col=0)
df
```

학생들의 키에대한 히스토그램을 그려보자

```python
df.plot(kind='hist',y='Height')
```

아무지정안하면 10개로 항목을 나뉘어 출력한다 15개로 나눠어 출력해보자

```python
df.plot(kind='hist', y='Height', bins=15)
```

더 상새하게보자

```python
df.plot(kind='hist', y='Height', bins=200)
```
