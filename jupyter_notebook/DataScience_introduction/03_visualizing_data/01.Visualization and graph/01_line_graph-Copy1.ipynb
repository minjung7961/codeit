```python
%matplotlib inline
```

```python
import pandas as pd
```

```python
df = pd.read_csv('data/broadcast.csv', index_col=0)
df
```

그래프를 써보자

```python
df.plot(kind = 'line') # 근데 기본이 선그래프라서 파라미터 보내는것은 생략해도 나옴
```

x 축엔 rows 값이나오고 y축엔 columns 값이 나온다

만약 한방송사만보고싶다면 어케해야할까?

```python
df.plot(y='KBS')
```

두개 보고싶으면? 방법2개

```python
df.plot(y=['KBS','MBC'])
```

```python
df[['KBS','JTBC']]
```

```python
df[['KBS','JTBC']].plot()
```
